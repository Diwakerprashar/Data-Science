```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
colours = ['Red','Green','yellow','Purple','Black']
```

```python
df = pd.read_csv('weatherAUS.csv')
df.head()
```

```python
g = df.groupby('Location')[['MinTemp','MaxTemp','Rainfall']].mean()
```

```python
df.groupby('Location')[['MinTemp','MaxTemp','Rainfall','WindGustSpeed']].mean()
```

```python
Rmax = g['Rainfall'].nlargest(5)
print(Rmax.index)
print(Rmax.values)
```

```python
fig = plt.figure(figsize = [8,6])
axes = fig.add_axes([0.1,0.1,1,0.8])
axes.bar(Rmax.index,Rmax.values,width = 0.6,color = colours)
plt.title('Maximum Rainfall Locations',fontsize = 20)
plt.xlabel('Locations',fontsize=20)
plt.ylabel('Max Rainfall.',fontsize = 20)
```

```python
Rmin = g['Rainfall'].nsmallest(5)
print(Rmin.index)
print(Rmin.values)
```

```python
fig = plt.figure(figsize = [8,6])
axes = fig.add_axes([0.1,0.1,1,0.8])
axes.bar(Rmin.index,Rmin.values,width = 0.6,color = colours)
plt.title('Minimum Rainfall Locations',fontsize = 20)
plt.xlabel('Locations',fontsize=20)
plt.ylabel('Min Rainfall.',fontsize = 20)
```

```python
T = g['MaxTemp'].nlargest(5)
print(T.index,T.values)
```

```python
fig = plt.figure(figsize = [8,6])
axes = fig.add_axes([0.1,0.1,1,0.8])
axes.bar(T.index,T.values,width = 0.6,color = colours)
plt.title('Maximum Temperature Locations',fontsize = 20)
plt.xlabel('Locations',fontsize=20)
plt.ylabel('Max Temp.',fontsize = 20)
```

```python
print('Max Temp',T.values)
print('Max Rain',Rmax.values)
print(T)
print(Rmax)
```
