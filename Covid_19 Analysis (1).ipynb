```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
colours = ['Red','Green','yellow','cyan','orange']
```

```python
df = pd.read_csv("Covid19Worldometer_data.csv")
df.head(5)
```

```python
x1 = df.groupby('Continent')['TotalCases'].sum()
x2 = df.groupby('Continent')['TotalDeaths'].sum()
x3 = df.groupby('Continent')['TotalRecovered'].sum()
x1   ,x2   ,x3
y = ['Africa','Asia','Australia/Oceania','Europe','North America','South America']
```

```python
fig = plt.figure(figsize=(10,6))
axes = fig.add_axes([0.1,1,1,0.8])
plt.title('Covid Cases',fontsize=30)
axes.set_xlabel('Continents',fontsize=20)
axes.set_ylabel('Covid 19 cases',fontsize=20)
axes.bar(y,x1,color=colours)
```

```python
fig = plt.figure(figsize=(10,6))
axes = fig.add_axes([0.1,1,1,0.8])
plt.title('Covid Cases',fontsize=30)
axes.set_xlabel('Continents',fontsize=20)
axes.set_ylabel('Deaths',fontsize=20)
axes.bar(y,x2,color=colours)
```

```python
fig = plt.figure(figsize=(10,6))
axes = fig.add_axes([0.1,1,1,0.8])
plt.title('Covid Cases',fontsize=30)
axes.set_xlabel('Continents',fontsize=20)
axes.set_ylabel('Recovered',fontsize=20)
axes.bar(y,x3,color=colours)
```

```python
mean_cases = df.groupby('Continent')['Tot Cases/1M pop'].mean()
mean_death = df.groupby('Continent')['Deaths/1M pop'].mean()
mean_tests = df.groupby('Continent')['Tests/1M pop'].mean()
```

```python
fig = plt.figure(figsize=(10,6))
axes = fig.add_axes([0.1,1,1,0.8])
plt.title('Average cases Per Million Population',fontsize=30)
axes.set_xlabel('Continents',fontsize=20)
axes.set_ylabel('Cases Per Million',fontsize=20)
axes.bar(y,mean_cases,color=colours)
```

```python
fig = plt.figure(figsize=(10,6))
axes = fig.add_axes([0.1,1,1,0.8])
plt.title('Average Death Per Million Population',fontsize=30)
axes.set_xlabel('Continents',fontsize=20)
axes.set_ylabel('Deaths',fontsize=20)
axes.bar(y,mean_death,color=colours)
```

```python
fig = plt.figure(figsize=(10,6))
axes = fig.add_axes([0.1,1,1,0.8])
plt.title('Average No. of Tests Per Million Population',fontsize=30)
axes.set_xlabel('Continents',fontsize=20)
axes.set_ylabel('Avg Tests',fontsize=20)
axes.bar(y,mean_tests,color=colours)
```
